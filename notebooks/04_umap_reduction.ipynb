# 04. UMAP Dimensionality Reduction

This notebook applies **UMAP** to the feature representations generated in the vectorisation stage.

## Purpose
The aim of this notebook is to:
- load the high-dimensional **TF-IDF** representation
- load the dense **S-BERT** embeddings
- reduce both representations to **10 dimensions**
- save the reduced outputs for downstream clustering

## Why UMAP?
UMAP was selected because it preserves local structure effectively while producing lower-dimensional representations that are more suitable for clustering and visualisation than the original high-dimensional feature spaces.

## Expected inputs
This notebook expects the feature files produced in the vectorisation stage:

```python
combined_tfidf_matrix.pkl
sbert_embeddings.npy
```

## Expected outputs
The notebook produces:
- `tfidf_umap_10d.npy`
- `sbert_umap_10d.npy`




## 1. Import libraries

This section imports the required libraries for dimensionality reduction and file loading.


````python
import umap
import numpy as np
import pickle
from pathlib import Path
````

## 2. Define input and output paths

**Note:** The input feature files are not included in the public repository due to size and data availability constraints.


````python
tfidf_input_file = Path("combined_tfidf_matrix.pkl")
sbert_input_file = Path("sbert_embeddings.npy")

tfidf_output_file = Path("tfidf_umap_10d.npy")
sbert_output_file = Path("sbert_umap_10d.npy")
````

## 3. Load TF-IDF features

The TF-IDF matrix is loaded from a pickle file created during vectorisation.


````python
with open(tfidf_input_file, "rb") as f:
    tfidf_matrix = pickle.load(f)

print("Loaded TF-IDF matrix shape:", tfidf_matrix.shape)
````

## 4. Reduce TF-IDF matrix to 10 dimensions with UMAP

The TF-IDF representation is reduced to a lower-dimensional space using cosine distance, which is well suited to sparse text data.


````python
tfidf_umap_model = umap.UMAP(
    n_components=10,
    metric="cosine",
    random_state=42
)

tfidf_umap_10d = tfidf_umap_model.fit_transform(tfidf_matrix)

print("TF-IDF UMAP-reduced shape:", tfidf_umap_10d.shape)
````

## 5. Save TF-IDF UMAP output

````python
np.save(tfidf_output_file, tfidf_umap_10d)

print(f"Saved: {tfidf_output_file}")
````

## 6. Load S-BERT embeddings

The S-BERT embeddings are loaded from the `.npy` file created during vectorisation.


````python
sbert_embeddings = np.load(sbert_input_file)

print("Loaded S-BERT embedding shape:", sbert_embeddings.shape)
````

## 7. Reduce S-BERT embeddings to 10 dimensions with UMAP

Cosine distance is also used here because it is commonly effective for semantic embedding spaces.


````python
sbert_umap_model = umap.UMAP(
    n_components=10,
    metric="cosine",
    random_state=42
)

sbert_umap_10d = sbert_umap_model.fit_transform(sbert_embeddings)

print("S-BERT UMAP-reduced shape:", sbert_umap_10d.shape)
````

## 8. Save S-BERT UMAP output

````python
np.save(sbert_output_file, sbert_umap_10d)

print(f"Saved: {sbert_output_file}")
````

## 9. Summary

At this point:
- the **TF-IDF** representation has been reduced from a sparse high-dimensional space to **10 dimensions**
- the **S-BERT** representation has been reduced from **384 dimensions** to **10 dimensions**

These reduced representations are used in the next stage for clustering experiments with HDBSCAN, K-Means, and BIRCH.
