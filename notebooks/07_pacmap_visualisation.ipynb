# 07. PaCMAP Visualisation

This notebook visualises the clustering results in **two dimensions** using **PaCMAP**.

## Purpose
The aim of this notebook is to:
- load the reduced feature representations and predicted cluster labels
- project the clustered data into 2D space using **PaCMAP**
- visualise the six final experiments consistently
- support qualitative interpretation of cluster structure

## Why PaCMAP?
PaCMAP was selected for visualisation because it produced clearer and more interpretable 2D layouts than t-SNE and UMAP during experimentation. This made it more suitable for comparing cluster separation and compactness across the six final experiments.

## Experimental note
Several visualisation approaches were explored. After comparison, PaCMAP was retained here because it offered the most useful balance between local grouping and global structure in the final visual analysis.

## Expected inputs
This notebook expects:
- reduced representations:
  - `tfidf_umap_10d.npy`
  - `sbert_umap_10d.npy`
- cluster labels:
  - `hdbscan_tfidf_labels.npy`
  - `hdbscan_sbert_labels.npy`
  - `kmeans_tfidf_labels.npy`
  - `kmeans_sbert_labels.npy`
  - `birch_tfidf_labels.npy`
  - `birch_sbert_labels.npy`




## 1. Import libraries

This section imports the required libraries for loading arrays and creating 2D visualisations.


```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pacmap
```

## 2. Define input files

```python
tfidf_data_file = Path("tfidf_umap_10d.npy")
sbert_data_file = Path("sbert_umap_10d.npy")

hdbscan_tfidf_file = Path("hdbscan_tfidf_labels.npy")
hdbscan_sbert_file = Path("hdbscan_sbert_labels.npy")

kmeans_tfidf_file = Path("kmeans_tfidf_labels.npy")
kmeans_sbert_file = Path("kmeans_sbert_labels.npy")

birch_tfidf_file = Path("birch_tfidf_labels.npy")
birch_sbert_file = Path("birch_sbert_labels.npy")
```

## 3. Load reduced representations and labels

```python
tfidf_umap_10d = np.load(tfidf_data_file)
sbert_umap_10d = np.load(sbert_data_file)

hdbscan_tfidf_labels = np.load(hdbscan_tfidf_file)
hdbscan_sbert_labels = np.load(hdbscan_sbert_file)

kmeans_tfidf_labels = np.load(kmeans_tfidf_file)
kmeans_sbert_labels = np.load(kmeans_sbert_file)

birch_tfidf_labels = np.load(birch_tfidf_file)
birch_sbert_labels = np.load(birch_sbert_file)

print("Loaded all representations and labels successfully.")
```

## 4. Define PaCMAP helper function

A reusable function is used to project a representation into 2D and visualise it with the corresponding cluster labels.


```python
def plot_pacmap_projection(X, labels, title, random_state=42):
    reducer = pacmap.PaCMAP(n_components=2, random_state=random_state)
    X_2d = reducer.fit_transform(X)

    plt.figure(figsize=(8, 6))
    plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, s=4, alpha=0.7)
    plt.title(title)
    plt.xlabel("PaCMAP Dimension 1")
    plt.ylabel("PaCMAP Dimension 2")
    plt.tight_layout()
    plt.show()

    return X_2d
```

## 5. HDBSCAN + TF-IDF visualisation

```python
hdbscan_tfidf_2d = plot_pacmap_projection(
    tfidf_umap_10d,
    hdbscan_tfidf_labels,
    "HDBSCAN + TF-IDF"
)
```

## 6. HDBSCAN + S-BERT visualisation

```python
hdbscan_sbert_2d = plot_pacmap_projection(
    sbert_umap_10d,
    hdbscan_sbert_labels,
    "HDBSCAN + S-BERT"
)
```

## 7. K-Means + TF-IDF visualisation

```python
kmeans_tfidf_2d = plot_pacmap_projection(
    tfidf_umap_10d,
    kmeans_tfidf_labels,
    "K-Means + TF-IDF"
)
```

## 8. K-Means + S-BERT visualisation

```python
kmeans_sbert_2d = plot_pacmap_projection(
    sbert_umap_10d,
    kmeans_sbert_labels,
    "K-Means + S-BERT"
)
```

## 9. BIRCH + TF-IDF visualisation

```python
birch_tfidf_2d = plot_pacmap_projection(
    tfidf_umap_10d,
    birch_tfidf_labels,
    "BIRCH + TF-IDF"
)
```

## 10. BIRCH + S-BERT visualisation

```python
birch_sbert_2d = plot_pacmap_projection(
    sbert_umap_10d,
    birch_sbert_labels,
    "BIRCH + S-BERT"
)
```

## 11. Final summary

At this stage, the clustering outputs have been visualised in 2D using PaCMAP. This supports the final interpretation of:
- cluster separation
- cluster compactness
