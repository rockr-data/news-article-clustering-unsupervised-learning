# 05. Clustering Experiments

This notebook runs the six clustering experiments used in the dissertation by combining:
- **two text representations**: TF-IDF and S-BERT
- **three clustering algorithms**: HDBSCAN, K-Means, and BIRCH

## Purpose
The aim of this notebook is to:
- load the UMAP-reduced TF-IDF and S-BERT representations
- run six clustering experiments in a consistent way
- store the predicted cluster labels for later evaluation and interpretation

## Experimental note
The six experiments were developed and tested separately during the  workflow. After trying multiple parameter settings, the selected parameters included here are the final ones used for the reported experiments.

## Expected inputs
This notebook expects the reduced files produced during the UMAP stage:

```python
tfidf_umap_10d.npy
sbert_umap_10d.npy
```

## Expected outputs
The notebook produces cluster label files for all six experiments:

- `hdbscan_tfidf_labels.npy`
- `hdbscan_sbert_labels.npy`
- `kmeans_tfidf_labels.npy`
- `kmeans_sbert_labels.npy`
- `birch_tfidf_labels.npy`
- `birch_sbert_labels.npy`



## 1. Import libraries

This section imports the required libraries for clustering and file handling.


````python
import numpy as np
from pathlib import Path

import hdbscan
from sklearn.cluster import KMeans, Birch
````

## 2. Define input and output files




````python
tfidf_input_file = Path("tfidf_umap_10d.npy")
sbert_input_file = Path("sbert_umap_10d.npy")

hdbscan_tfidf_output = Path("hdbscan_tfidf_labels.npy")
hdbscan_sbert_output = Path("hdbscan_sbert_labels.npy")

kmeans_tfidf_output = Path("kmeans_tfidf_labels.npy")
kmeans_sbert_output = Path("kmeans_sbert_labels.npy")

birch_tfidf_output = Path("birch_tfidf_labels.npy")
birch_sbert_output = Path("birch_sbert_labels.npy")
````

## 3. Load reduced representations

````python
tfidf_umap_10d = np.load(tfidf_input_file)
sbert_umap_10d = np.load(sbert_input_file)

print("TF-IDF reduced shape:", tfidf_umap_10d.shape)
print("S-BERT reduced shape:", sbert_umap_10d.shape)
````

## 4. Define helper function for cluster summary

This function prints a compact summary of the resulting labels, including the number of clusters and the number of noise points where applicable.


````python
def summarise_labels(labels, model_name):
    unique_labels = np.unique(labels)
    n_clusters = len(unique_labels[unique_labels != -1]) if -1 in unique_labels else len(unique_labels)
    n_noise = int((labels == -1).sum()) if -1 in unique_labels else 0

    print(f"{model_name}")
    print(f"Number of clusters: {n_clusters}")
    print(f"Number of noise points: {n_noise}")
    print("-" * 50)
````

## 5. HDBSCAN on UMAP-reduced TF-IDF

The final HDBSCAN configuration below reflects the selected settings after parameter testing.


````python
hdbscan_tfidf_model = hdbscan.HDBSCAN(
    min_cluster_size=50,
    min_samples=None,
    metric="euclidean",
    cluster_selection_method="eom"
)

hdbscan_tfidf_labels = hdbscan_tfidf_model.fit_predict(tfidf_umap_10d)
summarise_labels(hdbscan_tfidf_labels, "HDBSCAN on TF-IDF")
````

## 6. Save HDBSCAN TF-IDF labels

````python
np.save(hdbscan_tfidf_output, hdbscan_tfidf_labels)
print(f"Saved: {hdbscan_tfidf_output}")
````

## 7. HDBSCAN on UMAP-reduced S-BERT

The final HDBSCAN configuration below reflects the selected settings after parameter testing.


````python
hdbscan_sbert_model = hdbscan.HDBSCAN(
    min_cluster_size=50,
    min_samples=None,
    metric="euclidean",
    cluster_selection_method="eom"
)

hdbscan_sbert_labels = hdbscan_sbert_model.fit_predict(sbert_umap_10d)
summarise_labels(hdbscan_sbert_labels, "HDBSCAN on S-BERT")
````

## 8. Save HDBSCAN S-BERT labels

````python
np.save(hdbscan_sbert_output, hdbscan_sbert_labels)
print(f"Saved: {hdbscan_sbert_output}")
````

## 9. K-Means on UMAP-reduced TF-IDF

The K-Means configuration below reflects the selected settings after parameter testing.


````python
kmeans_tfidf_model = KMeans(
    n_clusters=5,
    random_state=42,
    n_init=10
)

kmeans_tfidf_labels = kmeans_tfidf_model.fit_predict(tfidf_umap_10d)
summarise_labels(kmeans_tfidf_labels, "K-Means on TF-IDF")
````

## 10. Save K-Means TF-IDF labels

````python
np.save(kmeans_tfidf_output, kmeans_tfidf_labels)
print(f"Saved: {kmeans_tfidf_output}")
````

## 11. K-Means on UMAP-reduced S-BERT

The K-Means configuration below reflects the selected settings after parameter testing.


````python
kmeans_sbert_model = KMeans(
    n_clusters=5,
    random_state=42,
    n_init=10
)

kmeans_sbert_labels = kmeans_sbert_model.fit_predict(sbert_umap_10d)
summarise_labels(kmeans_sbert_labels, "K-Means on S-BERT")
````

## 12. Save K-Means S-BERT labels

````python
np.save(kmeans_sbert_output, kmeans_sbert_labels)
print(f"Saved: {kmeans_sbert_output}")
````

## 13. BIRCH on UMAP-reduced TF-IDF

The BIRCH configuration below reflects the selected settings after parameter testing.


````python
birch_tfidf_model = Birch(
    n_clusters=5,
    threshold=0.5,
    branching_factor=50
)

birch_tfidf_labels = birch_tfidf_model.fit_predict(tfidf_umap_10d)
summarise_labels(birch_tfidf_labels, "BIRCH on TF-IDF")
````

## 14. Save BIRCH TF-IDF labels

````python
np.save(birch_tfidf_output, birch_tfidf_labels)
print(f"Saved: {birch_tfidf_output}")
````

## 15. BIRCH on UMAP-reduced S-BERT

The BIRCH configuration below reflects the selected settings after parameter testing.


````python
birch_sbert_model = Birch(
    n_clusters=5,
    threshold=0.5,
    branching_factor=50
)

birch_sbert_labels = birch_sbert_model.fit_predict(sbert_umap_10d)
summarise_labels(birch_sbert_labels, "BIRCH on S-BERT")
````

## 16. Save BIRCH S-BERT labels

````python
np.save(birch_sbert_output, birch_sbert_labels)
print(f"Saved: {birch_sbert_output}")
````
